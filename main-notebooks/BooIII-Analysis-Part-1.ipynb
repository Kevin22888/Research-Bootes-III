# Bootes III: Finding core members

Note that we convert celestial coordinates $(\alpha, \delta)$ to $(\Delta \alpha, \Delta \delta)$

1. Identify BooIII spectroscopic member stars from S5 data
    - Spatial cut with radius of 3 degrees centred on BooIII.
    - Filter out poor measurement stars
    - RV and PM cuts
    - Crossmatch with RRL catalogue to find RRL stars
    - Take only the stars within 5 half-light radius
    - Plot and compare with data from Carlin & Sand 2018

    
2. Find kinematics and metallicity of BooIII from the members
    - Compute systemic velocity and its dispersion (excluding RRLs)
    - Compute the proper motion and metallicity.


3. Use these systemic information to identify possible left out members.
    - Consider uncertainties of the data points (rather than an ordinary box cut without error bars)

### Preparing the data from Carlin & Sand 2018

````python
# The cell is just processing a table
table_carlin18 = ascii.read('carlin2018_members.txt', header_start=2, data_start=4, data_end=-5, delimiter='\t')
table_carlin18 = table_carlin18[table_carlin18['Member?'] == 'Y']

table_carlin18.rename_column('alpha', 'ra')
table_carlin18.rename_column('delta', 'dec')
table_carlin18.rename_column('v_helio', 'rv')

def separate_errors(table, column_name):
    """An exclusive helper function for a specific table. One-time use only.
    """
    nominals = []
    errors = []

    for rv in table[column_name]:
        nominal, error = rv.split(' +or- ')
        nominals.append(float(nominal))
        errors.append(float(error))

    table[column_name] = nominals
    table[column_name + '_error'] = errors

separate_errors(table_carlin18, 'rv')

fehs = []

for feh in table_carlin18['[Fe/H]^a']:
    if feh == 'cdots':
        fehs.append(np.nan)
    else:
        fehs.append(float(feh))

table_carlin18['feh'] = fehs
````
